Removing Outliers

```python
# function to remove outliers using the IQR method
def remove_outliers(df,column):
    Q1= df[column].quantile(0.25)
    Q3= df[column].quantile(0.75)
    IQR=Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR
    df=df[(df[column]>=lower_limit) & (df[column]<=upper_limit)]
    return df
```

```python
#apply the IQR method to continous column
continous_column=['quantity_tons_log', 'thickness_log', 'width','selling_price_log']
for col in continous_column:
    df1=remove_outliers(df1,col) 
```

Classification model which predict the status column Won/Lost

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,auc,roc_curve,confusion_matrix,classification_report

from imblearn.combine import SMOTETomek

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, ExtraTreesClassifier
from xgboost import XGBClassifier
```

```python
df_class=df1.copy()
```

```python
df_class.info()
```

```python
#filter the status column, we need only won and lost values
df_class_status= df_class[(df_class['status']==1) | (df_class['status']==0)]
```

```python
df_class_status
```

```python
df_class_status["status"].unique()
```

```python
df_class_status["status"].value_counts()
```

Using SMOTETomek for oversampling the data since the Target column is imbalanced which will affect the model accuarcy

```python
df_class_status.columns
```

```python
#independent column
x=df_class_status.drop(columns=['item_date','status','delivery date'],axis=1)
#dependent column
y= df_class_status['status']
```

```python
x.shape,y.shape
```

```python
y.value_counts()
```

Oversampling the data with SMOTETomek

```python
x_new,y_new=SMOTETomek().fit_resample(x,y)
```

```python
x_new.shape,y_new.shape
```

```python
y_new.value_counts()
```

Now the Target column is balanced hence preparing ML model 

```python
def best_algorithm_class(x_data,y_data,algorithm):
    x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=42)

    model= algorithm().fit(x_train,y_train)
    y_pred=model.predict(x_test)

    model_accuracy=accuracy_score(y_test,y_pred)

    metrics={"Algorithm":algorithm.__name__,"Accuracy_Score":model_accuracy}
    return metrics
```

```python
print(best_algorithm_class(x_new,y_new,DecisionTreeClassifier))
print(best_algorithm_class(x_new,y_new,RandomForestClassifier))
print(best_algorithm_class(x_new,y_new,GradientBoostingClassifier))
print(best_algorithm_class(x_new,y_new,AdaBoostClassifier))
print(best_algorithm_class(x_new,y_new,ExtraTreesClassifier))
print(best_algorithm_class(x_new,y_new,XGBClassifier))
```

After SMOTE-Tomek, ExtraTreesClassifier has the best accuracy.

Tuning the hyperparameter by GridsearchCV

```python
x_train, x_test, y_train, y_test = train_test_split(x_new, y_new, test_size = 0.20, random_state = 42)

parameter={'n_estimators':[50,100,200],'max_depth':[10,20,30],'min_samples_split':[2,4,6],'min_samples_leaf':[1,2,3]}

grid_search=GridSearchCV(estimator=ExtraTreesClassifier(random_state=42),param_grid=parameter,cv=5, scoring="accuracy",n_jobs=-1)

grid_search.fit(x_train,y_train)
print("Best Score: ", grid_search.best_score_,"Best Parameters: ", grid_search.best_params_)
```

passing the best hyperparameters in the ExtraTreesClassifier algorithm

```python
x_train, x_test, y_train, y_test = train_test_split(x_new, y_new, test_size = 0.20, random_state = 42)
model1=ExtraTreesClassifier(max_depth= 30,min_samples_leaf= 1, min_samples_split= 2,n_estimators= 200).fit(x_train,y_train)
y_predict=model1.predict(x_test)
# Evaluation Metrics:
print("Accuracy Score: {:.4f}".format(accuracy_score(y_test, y_predict)))
print("Precision Score: {:.4f}".format(precision_score(y_test, y_predict)))
print("Recall Score: {:.4f}".format(recall_score(y_test, y_predict)))
print("F1 Score: {:.4f}".format(f1_score(y_test, y_predict)))
```

```python
#confusion matrix
print("Confusion matrix:")
print(confusion_matrix(y_test,y_predict))
```

```python
# classification report
print("Classification_report:")
print(classification_report(y_test,y_predict))
```

```python
# Receiver Operating Characteristic (ROC) Curve
FP,TP,Threshold=roc_curve(y_test,y_predict)
print(FP)
print(TP)
print(Threshold)
```

```python
# Area Under the Curve (AUC)
auc_curve = auc(x=FP, y=TP)
print("auc_curve: ", auc_curve)
```

```python
#create a plot for ROC and AUC curve
roc_point={"ROC curve (area)":round(auc_curve,2)}
plt.plot(FP,TP,label=roc_point)
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.1])
plt.xlabel("False Positive")
plt.ylabel("True Positive")
plt.plot([0,1],[0,1],"k--")
plt.legend(loc="lower right")
plt.show()
```

Regression model which predict the selling_price

```python
#packages

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, ExtraTreesRegressor
from xgboost import XGBRegressor
```

```python
df_reg=df1.copy()
df_reg.head()
```

```python
#independent
x1=df_reg.drop(['item_date','delivery date','selling_price_log'],axis=1)
#dependent
y1=df_reg['selling_price_log']
```

```python
def best_algorithm_reg(df,algorithm):

    x=df_reg.drop(['item_date','delivery date','selling_price_log'],axis=1)
    y=df_reg['selling_price_log']

    x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.2,random_state=42)
    model2=algorithm().fit(x_train,y_train)
    y_pred=model2.predict(x_test)

    r2=r2_score(y_test,y_pred)

    metrics={'Alogrithm':algorithm.__name__,'R2 Score':r2}
    return metrics
```

```python
print(best_algorithm_reg(df_reg,LinearRegression))
print(best_algorithm_reg(df_reg,DecisionTreeRegressor))
print(best_algorithm_reg(df_reg,RandomForestRegressor))
print(best_algorithm_reg(df_reg,GradientBoostingRegressor))
print(best_algorithm_reg(df_reg,AdaBoostRegressor))
print(best_algorithm_reg(df_reg,ExtraTreesRegressor))
print(best_algorithm_reg(df_reg,XGBRegressor))
```

RandomForestRegressor has the best R2 Score

```python
x_train, x_test, y_train, y_test = train_test_split(x_new, y_new, test_size = 0.20, random_state = 42)
model2=RandomForestRegressor(n_estimators=100,max_depth=30,min_samples_split=2,min_samples_leaf=1).fit(x_train,y_train)
y_predict=model2.predict(x_test)
# Evaluation Metrics:
print("MAE: {:.4f}".format(mean_absolute_error(y_test,y_predict)))
print("MSE: {:.4f}".format(mean_squared_error(y_test,y_predict)))
print("RMSE: {:.4f}".format(np.sqrt(mean_squared_error(y_test,y_predict))))
print("R2: {:.4f}".format(r2_score(y_test,y_predict)))
```
